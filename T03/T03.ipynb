### NumPy<a class="anchor" id="numpy"></a>

NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a **multidimensional array object**, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.

At the core of the NumPy package, is the ndarray object. This encapsulates n-dimensional arrays of **homogeneous** data types, with many operations being performed in compiled code for performance. There are several important differences between NumPy arrays and the standard Python sequences:

- NumPy arrays have a **fixed size at creation**, unlike Python lists (which can grow dynamically). Changing the size of an ndarray will create a new array and delete the original.
-  The elements in a NumPy array are all required to be of the **same data type**, and thus will be the same size in memory. The exception: one can have **arrays of (Python, including NumPy) objects**, thereby allowing for arrays of different sized elements.
-  NumPy arrays facilitate advanced mathematical and other types of operations on large numbers of data. Typically, such operations are executed more efficiently and with less code than is possible using Python’s built-in sequences.
-  A growing plethora of scientific and mathematical Python-based packages are using NumPy arrays; though these typically support Python-sequence input, they convert such input to NumPy arrays prior to processing, and they often output NumPy arrays. In other words, in order to efficiently use much (perhaps even most) of today’s scientific/mathematical Python-based software, just knowing how to use Python’s built-in sequence types is insufficient - one also needs to know how to use NumPy arrays.

**Example of NumPy - Multiple Linear Regression**

To illustrate the usage of numpy in a statistical application, consider the problem of multiple linear regression where the response variable is explained by a linear combination of predictor variables. Now, a dataset of 442 diabetes patients is studied. In this data file, the first column stores the values of a response variable which is indeed a quantitative measure of disease progression one year after baseline and the remaining columns are the age, sex, body mass index, average blood pressure, and six blood serum measurements. All the observations are stored in the data file `diabetes.csv` (Note that this file has no header information).

To read the observations into a `numpy.ndarray`, the function `np.genfromtxt()` can be used here.

````python
data = np.genfromtxt('diabetes.csv', delimiter=',')
````

````python
data.shape # dimension of the array
````

````python
Y = data[:,0]   # extract the target and input variables
X = data[:,1:]
Xstar = np.concatenate((np.ones((data.shape[0],1)), X), axis=1)
````

#### Compute the regression coefficients

````python
bcoeff = np.linalg.inv(np.dot(Xstar.T,Xstar)) @ np.dot(Xstar.T,Y)
bcoeff
````

#### Compute the standard errors of estimated coeffcients and hence the t-ratios

````python
# predicted values of target
pred = np.dot(Xstar,bcoeff)
# residuals
resid = Y - pred
# estimated variance of errors
sig2 = np.dot(resid,resid) / (data.shape[0]-data.shape[1]-1)
# estimated covariance matrix of estimated coefficients
cov = sig2 * np.linalg.inv(np.dot(Xstar.T,Xstar))
# estimated standard error of estimated coefficients
bse = np.sqrt(np.diag(cov))
# t-ratio
tratio = bcoeff / bse
print(bse)
print(tratio)
````

#### Compute the goodness of fit measure, R-square

````python
Rsq = 1 - np.dot(resid,resid) / (np.var(Y)*data.shape[0])
Rsq
````

#### Display the fitted result graphically (`matplotlib`)

````python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(Y, pred, '.')

# plot a line, a perfit predict would all fall on this line
x = np.linspace(0, 350, 100)
y = x
plt.plot(x, y)
````

#### `optimize` subpackage

Optimization is a typical task in statistics, e.g. maximum likelihood estimations of logistic regression models, non-linear regression models, estimation of non-parametric regression models. Actually, this subpackage contains a number of **unconstrained minimization algorithms** (Nelder-Mead Simplex, Broyden-Fletcher-Goldfarb-Shanno(BFGS), Newton-Conjugate-Gradient, Trust-Region Newton-Conjugate-Gradient, Trust-Region Truncated Generalized Lanczos / Conjugate Gradient, Trust-Region Nearly Exact, ), **constrained minimization algorithms** (Trust-Region Constrained, Sequential Least SQuares Programming and Least-squares), including **root-finding algorithm**.

The following example illustrates the analysis of kinetic data for an enzyme reaction which was considered by Averick et al. (1992) and this problem was formulated by Kowalik and Morrison (1968) originally. The residuals of the estimated model are given by

\begin{align}
    f_i (\boldsymbol{x}) & = \dfrac{x_0 (u_i^2 + u_i x_1)}{u_i^2 + u_i x_2 + x_3} - y_i , \hspace{5mm} i = 1,2,\ldots,11 \\
\end{align}

where $y_i$'s and $u_i$'s are the measurements and predictors respectively, $\boldsymbol{x} = x_1 , x_2 ,x_3)^{T}$ collects the parameters selected to minimize the sum of squared errors.

In order to improve the performance of numerical minimization of sum of squared errors, the gradient information, in the analytical form of Jacobian, is also provided.

\begin{eqnarray*}
J_{i0} & = & \dfrac{\partial f_i}{\partial {x_0}} = \dfrac{u_i^2 + u_i x_1}{u_i^2 + u_i x_2 + x_3} \\
J_{i1} & = & \dfrac{\partial f_i}{\partial {x_1}} = \dfrac{u_i x_0}{u_i^2 + u_i x_2 + x_3} \\
J_{i2} & = & \dfrac{\partial f_i}{\partial {x_2}} = -\dfrac{x_0 (u_i^2 + u_i x_1) u_i}{(u_i^2 + u_i x_2 + x_3)^2} \\
J_{i3} & = & \dfrac{\partial f_i}{\partial {x_3}} = -\dfrac{x_0 (u_i^2 + u_i x_1)}{(u_i^2 + u_i x_2 + x_3)^2} \\
\end{eqnarray*}

To find a physically meaningful solution, avoid potential division by zero and assure convergence to the global minimum we impose constraints $0\le x_j \le 100$, $j=0,1,2,3$.

In the following Python code, the functions `model(x,u)` and `fun(x,u,y)` define the predicted values of $y_i$ and residuals respectively. The `numpy.ndarray` objects `u` and `y` store the the values of $u_i$'s and $y_i$'s respectively. Furthermore, the `numpy.ndarray` object `x0` stores the the initial values of $x_j$ for $j=0,1,2,3$.

````python
from scipy.optimize import least_squares

def model(x,u):
    return x[0] * (u ** 2 + x[1] * u) / (u ** 2 + x[2] * u + x[3])
def fun(x,u,y):
    return model(x, u) - y
def jac(x,u,y):
    J = np.empty((u.size, x.size))
    den = u ** 2 + x[2] * u + x[3]
    num = u ** 2 + x[1] * u
    J[:,0] = num / den
    J[:,1] = x[0] * u / den
    J[:,2] = -x[0] * num * u / den ** 2
    J[:,3] = -x[0] * num / den ** 2
    return J
u = np.array([4.0, 2.0, 1.0, 5.0e-1, 2.5e-1, 1.67e-1, 1.25e-1, 1.0e-1,
              8.33e-2, 7.14e-2, 6.25e-2])
y = np.array([1.957e-1, 1.947e-1, 1.735e-1, 1.6e-1, 8.44e-2, 6.27e-2,
              4.56e-2, 3.42e-2, 3.23e-2, 2.35e-2, 2.46e-2])
x0 = np.array([2.5, 3.9, 4.15, 3.9])
res = least_squares(fun, x0, jac=jac, bounds=(0,100), args=(u,y), verbose=1)
````

The selected values of $x_j$'s are stored in the object `res.x`.

````python
res.x
````

To display the fitted result, the module `matplotlib` is used here.

````python
import matplotlib.pyplot as plt

u_test = np.linspace(0,5)
y_test = model(res.x, u_test)
plt.plot(u, y, 'o', markersize=4, label='data')
plt.plot(u_test, y_test, label='fitted model')
plt.xlabel("u")
plt.ylabel("y")
plt.legend(loc='lower right')
````

**Example 2 - Prediction of Lower Faces of Olivetti Dataset by Multiple Machine Learning Methods**

Another example which is originated from an example of official documentation of `scikit-learn` module, prediction of the lower faces based on the upper faces is performed for dataset of Olivetti faces. In the dataset, there are 400 faces of size 64x64 pixels which has been quantized to 256 grey levels.  For some subjects, the images were taken at different times, varying the lighting, facial expressions (open / closed eyes, smiling / not smiling) and facial details (glasses / no glasses). All the images were taken against a dark homogeneous background with the subjects in an upright, frontal position (with tolerance for some side movement)

The `targets` variable in integer an from 0 to 39 indicating the identity of the person pictured; however, with only 10 examples per class.

To demonstrate the machine features of `scikit-learn`, the models of Extra-trees classifier, K-Nearest Neighbor, linear regression and ridge regression are selected.

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import fetch_olivetti_faces
from sklearn.utils.validation import check_random_state

from sklearn.ensemble import ExtraTreesRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RidgeCV

# Load the faces datasets
data = fetch_olivetti_faces()
targets = data.target

data.images.shape
````

Since the image data is imported into an 3D array, it is reshaped into an 2D array and each row contains all the pixels of an image.

````python
data = data.images.reshape((len(data.images), -1))
````

````python
data.shape
````

The images of first 30 persons are allocated to the training set while the remainings are allocated to test test.

````python
train = data[targets < 30] 
test = data[targets >= 30]
````

To reduce the computational time, 5 images are selected for testing purpose.

````python
# Test on a subset of people
n_faces = 5
# initialization for random number generation
rng = check_random_state(4)
face_ids = rng.randint(test.shape[0], size=(n_faces, )) # randint() is a numpy func indeed.
test = test[face_ids, :]
````

Construct the target and input variables in the training and test tests.

````python
n_pixels = data.shape[1]
# Upper half of the faces
x_train = train[:, :(n_pixels + 1) // 2]
# Lower half of the faces
y_train = train[:, n_pixels // 2:]
x_test = test[:, :(n_pixels + 1) // 2]
y_test = test[:, n_pixels // 2:]
````

Construct a dicionary object `ESTIMATORS` which stores the function handlers of all machine learning algorithms. Then, the predicted results of test set are stored in another dictionary object `y_test_predict`. 

````python
ESTIMATORS = {
    "Extra trees": ExtraTreesRegressor(n_estimators=10, max_features=32,
                                       random_state=0),
    "K-nn": KNeighborsRegressor(),
    "Linear regression": LinearRegression(),
    "Ridge": RidgeCV(),
}

y_test_predict = dict()
for name, estimator in ESTIMATORS.items():
    estimator.fit(x_train, y_train)
    y_test_predict[name] = estimator.predict(x_test)
````

Finally, the predicted results are displayed graphically.

````python
# Plot the completed faces
image_shape = (64, 64)

n_cols = 1 + len(ESTIMATORS)
plt.figure(figsize=(2. * n_cols, 2.26 * n_faces))
plt.suptitle("Face completion with multi-output estimators", size=16)

for i in range(n_faces):
    true_face = np.hstack((x_test[i], y_test[i]))

    if i:
        sub = plt.subplot(n_faces, n_cols, i * n_cols + 1)
    else:
        sub = plt.subplot(n_faces, n_cols, i * n_cols + 1,
                          title="true faces")

    sub.axis("off")
    sub.imshow(true_face.reshape(image_shape),
               cmap=plt.cm.gray,
               interpolation="nearest")

    for j, est in enumerate(sorted(ESTIMATORS)):
        completed_face = np.hstack((x_test[i], y_test_predict[est][i]))

        if i:
            sub = plt.subplot(n_faces, n_cols, i * n_cols + 2 + j)

        else:
            sub = plt.subplot(n_faces, n_cols, i * n_cols + 2 + j,
                              title=est)

        sub.axis("off")
        sub.imshow(completed_face.reshape(image_shape),
                   cmap=plt.cm.gray,
                   interpolation="nearest")
````

### Keras<a class="anchor" id="keras"></a>

Keras is a high-level neural networks API, written in Python and capable of running on top of TensorFlow, CNTK, or Theano. It was developed with a focus on enabling fast experimentation.

Basically, Keras ia a deep learning library which 

- Allows for easy and fast prototyping (through user friendliness, modularity, and extensibility).
- Supports both convolutional networks and recurrent networks, as well as combinations of the two.
- Runs seamlessly on CPU and GPU.

To install the Keras, enter the following command under Anaconda Command Prompt:

`conda install keras`

By default, the Keras will use TensorFlow as the backend. So, the tensorflow module and its visualization tool tensorboard are installed automatically.

The core data structure of Keras is a model to organize **layers**. The simplest type of model is the `Sequential` model, a linear stack of layers. To build a more complexed model, the Keras functional API can be used. For more information of the functional API, visit the following link:

https://keras.io/getting-started/functional-api-guide/

In particular, Keras provided several predefined layers, namely, (1) Core layers, Convolutional layers, Pooling layers, Locally-connected layers, Recurrent layers, Embedding layers, Merge layers, Advanced activiations layers, Normalization layers, Noise layers, Layer wrappers (such as bidirectional layer). Also, we are allowed can write our own Keras layers as well.

The detailed description of layers in Keras can be found at:

https://keras.io/layers/about-keras-layers/

To build a sequential model, basically, the following three steps are necessary: 

1. [Specifying the input shape](#shape), 

2. [Compilation](#compile) and 

3. [Training](#train).

#### **1. Specifying the input shape**<a class="anchor" id="shape"></a>

Typically, the functions `Sequential()`, `Dense()` and `Activation()` are loaded into memory:

````python
from keras.models import Sequential
from keras.layers import Dense, Activation
````

Then, the model can be specified in either form:

```python
model = Sequential([
    Dense(32, input_shape=(784,)),
    Activation('relu')
])
```

or 

```python
model = Sequential()
model.add(Dense(32, input_dim=784))
model.add(Activation('relu'))
```

In either specification of the model above, the core layer (by `Dense()`) and the activation layer (by `Activiation()`) are included in the `Sequential` model. Actually, the model needs to know what input shape it should expect. In other words, we have to tell how much data is input to the model. So, the first layer in `Sequential` model needs to receive the information about its input shape. Afterwards, the following layers can infer the shape of data automatically.

To specify the shape of input data in the core layer, the first argument is `batch_size=` which controls the number of samples in each mini-batch being passed to the first layer of the model. Its **maximum** is the number of all samples, which makes the gradient descent algorithm accurate, the loss will decrease towards the the minimum if the learning rate ismall enough, but iterations are slower. The **minimum** is 1, resulting in stochastic gradient descent algorithm: Fast but the direction of the gradient step is based only on one sample, the loss may jump around. `batch_size` allows to adjust between the two extremes: accurate gradient direction and fast iteration. Also, the maximum value for `batch_size` may be limited if the model + data set does not fit into the available (GPU) memory.

The second input argument is `input_shape=` and its value is specified as a tuple of integers. When the value of this argument is None, the number of any positive integer may be expected. Note that the size of batch is not specified in this argument.

If the values of both `batch_size=32` and `input_shape=(28, 28)`are passed to a layer, it will then expect every batch of inputs to have the batch shape `(32, 28, 28)`.

```python
model = Sequential()
model.add(Dense(32, input_shape=(28,28)))
```

#### **2. Compilation**<a class="anchor" id="compile"></a>

After the model is specified, the optimization method is specified for training of model parameters. Depending on the modelling objectives, different optimization algorithms and criteria, are specified.

For a multi-class classification problem,
```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```
For a binary classification problem,
```python
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```
For a prediction of continuous target problem, 
```python
model.compile(optimizer='rmsprop',
              loss='mse')
```
Furthermore, customized metrics is also supported in Keras.
```python
import keras.backend as K

def mean_pred(y_true, y_pred):
    return K.mean(y_pred)

model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy', mean_pred])
```

Refer to all the example specifications in the `.compile()` method, the argument `optimizer=` specifies the optimization algorithm to minimize the loss function; the argument `loss=` is the objective function being minimized; and  the argument `metrics=` is specified in the classification problem.

The detailed information of optimization algorithms and loss functions can be found at:

https://keras.io/optimizers/  and  https://keras.io/losses/

respectively.

#### **3. Training**<a class="anchor" id="train"></a>

Keras models are trained on NumPy arrays of input data and labels. For training a model, the function `fit()` will be used typically. The input arguement `epochs=` is the number of iterations to train the model. Usually, the larger the value of this input argument, the more accurate the reuslt even though longer computational time will be spent. In addition, the input argument `batch_size=` is the number of samples per gradient update.  The detailed usage of the `fit()` function can be found at

https://keras.io/models/sequential/

For binary classification,

````python
model = Sequential()
model.add(Dense(32, activation='relu', input_dim=100))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Generate synthetic data for demonstration
import numpy as np
data = np.random.random((1000, 100))
labels = np.random.randint(2, size=(1000, 1))

# Train the model, iterating on the data in batches of 32 samples
model.fit(data, labels, epochs=10, batch_size=32)
````

For multi-class classification,

````python
model = Sequential()
model.add(Dense(32, activation='relu', input_dim=100))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Generate synthetic data for demonstration
import numpy as np
data = np.random.random((1000, 100))
labels = np.random.randint(2, size=(1000, 1))

# Train the model, iterating on the data in batches of 32 samples
model.fit(data, labels, epochs=10, batch_size=32)
````

````python
from __future__ import print_function
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import backend as K

batch_size = 128
num_classes = 10
epochs = 12

# input image dimensions
img_rows, img_cols = 28, 28

# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = mnist.load_data()

if K.image_data_format() == 'channels_first':
    x_train = x_train.reshape(x_train.shape[0], 1, img_rows, img_cols)
    x_test = x_test.reshape(x_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
    x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')

# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=input_shape))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])

model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test))
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
````

**Example 1 - Multiple linear regression**

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# read the data file into a pandas DataFrame
diab = pd.read_csv("diabetes.csv", 
                   header=None, 
                   names=['target', 'age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6'])

# use the train_test_split() function in scikit-learn to split the data
from sklearn.model_selection import train_test_split
varlist = list(diab.columns)
x_train, x_test, y_train, y_test = train_test_split(diab.loc[:,varlist[1:]], diab[varlist[0]], test_size=0.3, random_state=0)
y_train = y_train.values.reshape(x_train.shape[0],1)
y_test = y_test.values.reshape(x_test.shape[0],1)
n_features = x_train.shape[1]
n_epochs = 100000

import tensorflow as tf

x = tf.placeholder(tf.float32, shape=[None,n_features])
y = tf.placeholder(tf.float32, shape=[None,1])

W = tf.Variable(tf.random_normal([n_features,1]))
b = tf.Variable(tf.random_normal([1]))

yhat = tf.matmul(x, W) + b

# use sum of squares as the loss function
loss = tf.reduce_mean(tf.square(y - yhat))
optimizer = tf.train.GradientDescentOptimizer(.01).minimize(loss)

sess = tf.InteractiveSession()
sess.run(tf.global_variables_initializer())

for i in range(n_epochs):
    sess.run(optimizer, feed_dict={x: x_train, y: y_train})
    if ((i+1)%10000==0 or i==0):
        print("Iteration %s: loss = %s" % (i+1, loss.eval(feed_dict={x: x_train, y: y_train})))
````

````python
# display the weights and bias
print("Weight: \n%s" % sess.run(W).T)
print("Bias: \n%s" % sess.run(b).T)
````

````python
# compute the predicted values of the target for the test set
y_pred = sess.run(yhat, feed_dict={x: x_test})
plt.plot(y_test, y_pred, '.')

# plot a line, a perfit predict would all fall on this line
x = np.linspace(0, 350, 100)
y = x
plt.plot(x, y)
````

**Example 2 - Classification of MNIST Dataset**

Another example is a classical example of deep learning, namely, classification of MNIST dataset by ConvNet model. Note that the settings of this model is slightly different from the LeNet model in Keras example.

````python
import tensorflow as tf

# Import MNIST data
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
````

````python
# Training Parameters
learning_rate = 0.001
num_epochs = 500
batch_size = 128
display_epoch = 10

# Network Parameters
num_input = 784 # MNIST data input (img shape: 28*28)
num_classes = 10 # MNIST total classes (0-9 digits)
dropout = 0.75 # Dropout, probability to keep units

# tf Graph input
X = tf.placeholder(tf.float32, [None, num_input])
Y = tf.placeholder(tf.float32, [None, num_classes])
keep_prob = tf.placeholder(tf.float32) # dropout (keep probability)
````

````python
# Construct some wrappers for simplicity
def conv2d(x, W, b, strides=1):
    # Conv2D wrapper, with bias and relu activation
    x = tf.nn.conv2d(x, W, strides=[1, strides, strides, 1], padding='SAME')
    x = tf.nn.bias_add(x, b)
    return tf.nn.relu(x)

def maxpool2d(x, k=2):
    # MaxPool2D wrapper
    return tf.nn.max_pool(x, ksize=[1, k, k, 1], strides=[1, k, k, 1],
                          padding='SAME')

# Construct model
def conv_net(x, weights, biases, dropout):
    # MNIST data input is a 1-D vector of 784 features (28*28 pixels)
    # Reshape to match picture format [Height x Width x Channel]
    # Tensor input become 4-D: [Batch Size, Height, Width, Channel]
    x = tf.reshape(x, shape=[-1, 28, 28, 1])

    # Convolution Layer
    conv1 = conv2d(x, weights['wc1'], biases['bc1'])
    # Max Pooling (down-sampling)
    conv1 = maxpool2d(conv1, k=2)

    # Convolution Layer
    conv2 = conv2d(conv1, weights['wc2'], biases['bc2'])
    # Max Pooling (down-sampling)
    conv2 = maxpool2d(conv2, k=2)

    # Fully connected layer
    # Reshape conv2 output to fit fully connected layer input
    fc1 = tf.reshape(conv2, [-1, weights['wd1'].get_shape().as_list()[0]])
    fc1 = tf.add(tf.matmul(fc1, weights['wd1']), biases['bd1'])
    fc1 = tf.nn.relu(fc1)
    # Apply Dropout
    fc1 = tf.nn.dropout(fc1, dropout)

    # Output, class prediction
    out = tf.add(tf.matmul(fc1, weights['out']), biases['out'])
    return out
````

````python
# Store layers weight & bias
weights = {
    # 5x5 conv, 1 input, 32 outputs
    'wc1': tf.Variable(tf.random_normal([5, 5, 1, 32])),
    # 5x5 conv, 32 inputs, 64 outputs
    'wc2': tf.Variable(tf.random_normal([5, 5, 32, 64])),
    # fully connected, 7*7*64 inputs, 1024 outputs
    'wd1': tf.Variable(tf.random_normal([7*7*64, 1024])),
    # 1024 inputs, 10 outputs (class prediction)
    'out': tf.Variable(tf.random_normal([1024, num_classes]))
}

biases = {
    'bc1': tf.Variable(tf.random_normal([32])),
    'bc2': tf.Variable(tf.random_normal([64])),
    'bd1': tf.Variable(tf.random_normal([1024])),
    'out': tf.Variable(tf.random_normal([num_classes]))
}

# Construct model
logits = conv_net(X, weights, biases, keep_prob)
prediction = tf.nn.softmax(logits)

# Define loss and optimizer
loss_op = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(
    logits=logits, labels=Y))
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
train_op = optimizer.minimize(loss_op)

# Evaluate model
correct_pred = tf.equal(tf.argmax(prediction, 1), tf.argmax(Y, 1))
accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))

# Initialize the variables (i.e. assign their default value)
init = tf.global_variables_initializer()
````

````python
# Start training
with tf.Session() as sess:

    # Run the initializer
    sess.run(init)

    for epoch in range(1, num_epochs+1):
        batch_x, batch_y = mnist.train.next_batch(batch_size)
        # Run optimization op (backprop)
        sess.run(train_op, feed_dict={X: batch_x, Y: batch_y, keep_prob: dropout})
        if epoch % display_epoch == 0 or epoch == 1:
            # Calculate batch loss and accuracy
            loss, acc = sess.run([loss_op, accuracy], feed_dict={X: batch_x,
                                                                 Y: batch_y,
                                                                 keep_prob: 1.0})
            print("Epoch " + str(epoch) + ", Minibatch Loss= " + \
                  "{:.4f}".format(loss) + ", Training Accuracy= " + \
                  "{:.3f}".format(acc))

    print("Optimization Finished!")

    # Calculate accuracy for 256 MNIST test images
    print("Testing Accuracy:", \
        sess.run(accuracy, feed_dict={X: mnist.test.images[:256],
                                      Y: mnist.test.labels[:256],
                                      keep_prob: 1.0}))
````
